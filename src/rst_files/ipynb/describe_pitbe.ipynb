## PItBEについて

本モジュール`PItBE`は量子アルゴリズムBlock-Encoding法（以下BE法）の実行用モジュールであり、
本モジュールを用いることで簡単にBE法を実行することができる。

このページではBE法の理論的背景や`PItBE`で実行可能なことについて解説する。
特にBE法の理論的背景については日本語で解説されたテキストは非常に少ないので、
本テキストを介してBE法についての見識を深めてほしい。\
なお、本テキストについては量子コンピュータについての基礎的な知識や、古典コンピュータ上で
実行可能な量子コンピュータシミュレータが実行可能であることを前提としている。
これらの知識が不十分である場合[このサイト](https://dojo.qulacs.org/ja/latest/index.html)
で知識を身につけることを推奨する。
量子コンピュータの仕組みからシミュレータの実行方法に至るまで、丁寧に解説されているので初学者に
適している。

`PItBE`に含まれる個別の関数についてのプログラミング面での
説明は`Contents:`の欄を参照してほしい。
